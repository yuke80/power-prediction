## 電力消費量ヒストグラム
ヒストグラムでデータの分布を見ます。このヒストグラムは、横軸は電力消費量に対するビン数（横軸の区間の分割数）<br>
を５０にしたものです。縦軸はその区間に対する度数（発生頻度）を示しています。<br>


```python
# coding: utf-8
import pandas as pd

years = [2017, 2018, 2019, 2020]

# 四国電力の電力消費量データを読み込み
ed = [pd.read_csv(
    'data/juyo_shikoku_{:d}.csv'.format(year),
    skiprows=3,
    names=['DATE', 'TIME', 'consumption'],
    parse_dates={'date_hour': ['DATE', 'TIME']},
    index_col = "date_hour")
    for year in years
]
elec_data = pd.concat(ed)

# -- 可視化 --
import matplotlib.pyplot as plt

# 画像のサイズを設定する
plt.figure(figsize=(10, 6))

# ヒストグラム生成
plt.hist(elec_data['consumption'], bins=50, color="gray")
plt.xlabel('Power consumption(*10000 kW)')
plt.ylabel('count')

# グラフ保存
# plt.savefig('power-comsumption-histgram.png')
plt.show()
```
