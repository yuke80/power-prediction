## 気温ヒストグラム
ヒストグラムで気温の発生頻度を見ます。横軸が気温、縦軸が発生頻度です。ビン数は電力消費量と同じで５０です。<br>
電力消費量と異なり、１０度近辺と２５度近辺に２つの山を持つ二峰性の分布になっています。

```python
# coding: utf-8
import pandas as pd

years = [2017, 2018, 2019, 2020]

# 複数年の気象データを読み込み
tp = [pd.read_csv(
    './data/takamatsu{:d}.csv'.format(year),
    skiprows=5,
    names=['DATETIME', 'temperature', 'temperature2', 'temperature3', 
           'sunhour1','sunhour2','sunhour3','sunhour4',
           'insolation','insolation2','insolation3',
           'humid','humid2','humid3'],
    parse_dates={'date_hour': ["DATETIME"]},
    index_col='date_hour')
    for year in years
]

tmp = pd.concat(tp)

# -- 可視化 --
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# ヒストグラム生成
plt.hist(tmp['temperature'], bins=50, color="gray")
plt.xlabel('Temperature(C degree)')
plt.ylabel('count')

# グラフ保存
#plt.savefig('temperature-histgram.png')
plt.show()
```
